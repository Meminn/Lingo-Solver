```python
import requests
from bs4 import BeautifulSoup

# Türk alfabesindeki harfler
alphabet = ["a", "b", "c", "ç", "d", "e", "f", "g", "ğ", "h", "ı", "i", "j", "k", "l", "m", "n", "o", "ö", "p", "r", "s", "ş", "t", "u", "ü", "v", "y", "z"]

# 4, 5 ve 6 harfli kelimeler için listeler oluştur
four_letter_words = []
five_letter_words = []
six_letter_words = []

for letter in alphabet:
    for word_length in [4, 5, 6]:
        url = f"https://www.kelimetre.com/{letter}-ile-baslayan-{word_length}-harfli-kelimeler"
        
        try:
            # Sayfayı iste
            response = requests.get(url)
            response.raise_for_status()
            soup = BeautifulSoup(response.text, 'html.parser')

            # card class'ını bul
            cards = soup.find_all(class_="card")

            for card in cards:
                # card içindeki <ul> elementlerini bul
                ul_elements = card.find_all("ul")
                for ul in ul_elements:
                    # <ul> içindeki tüm <li> elementlerini bul
                    li_elements = ul.find_all("li")
                    # <li> içindeki title özniteliğini al
                    for li in li_elements:
                        # <li> içindeki <a> elementini bul
                        a_tag = li.find("a")
                        if a_tag:
                            # <a> elementinin title özniteliğini al
                            title = a_tag.get("title")
                            if title:
                                # Kelimeyi al ve gerekli işlemi yap
                                word = title.split()[0]  # "ücra nedir ?" -> "ücra"
                                if word_length == 4:
                                    four_letter_words.append(word)
                                elif word_length == 5:
                                    five_letter_words.append(word)
                                elif word_length == 6:
                                    six_letter_words.append(word)


        except requests.exceptions.RequestException as e:
            print(f"Error fetching URL {url}: {e}")

# Sonuçları yazdır
print("4 harfli kelimelerden ilk 10:", four_letter_words[:10])
print("5 harfli kelimelerden ilk 10:", five_letter_words[:10])
print("6 harfli kelimelerden ilk 10:", six_letter_words[:10])

# İsteğe bağlı: Sonuçları dosyalara kaydet
with open("four_letter_words.txt", "w", encoding="utf-8") as file:
    for word in four_letter_words:
        file.write(word + "\n")

with open("five_letter_words.txt", "w", encoding="utf-8") as file:
    for word in five_letter_words:
        file.write(word + "\n")

with open("six_letter_words.txt", "w", encoding="utf-8") as file:
    for word in six_letter_words:
        file.write(word + "\n")
```

```python
import requests
from bs4 import BeautifulSoup

# Türk alfabesi
alphabet = ["a", "b", "c", "ç", "d", "e", "f", "g", "ğ", "h", "ı", "i", "j", "k", "l", "m", "n", "o", "ö", "p", "r", "s", "ş", "t", "u", "ü", "v", "y", "z"]

# Kelimeler için dinamik sözlük
word_collections = {4: [], 5: [], 6: []}

for letter in alphabet:
    for word_length in word_collections.keys():
        url = f"https://www.kelimetre.com/{letter}-ile-baslayan-{word_length}-harfli-kelimeler"
        try:
            response = requests.get(url,  headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    })
            response.raise_for_status()
            soup = BeautifulSoup(response.text, 'html.parser')
            cards = soup.find_all(class_="card")

            for card in cards:
                ul_elements = card.find_all("ul")
                for ul in ul_elements:
                    li_elements = ul.find_all("li")
                    for li in li_elements:
                        title = li.get("title")
                        if title:
                            word_collections[word_length].append(title.strip())

        except requests.exceptions.RequestException as e:
            print(f"Error fetching URL {url}: {e}")

# Sonuçları yazdır
for length, words in word_collections.items():
    print(f"{length} harfli kelimelerden ilk 10:", words[:10])

# Dosyalara kaydet
for length, words in word_collections.items():
    with open(f"{length}_letter_words.txt", "w", encoding="utf-8") as file:
        file.write("\n".join(words))
```
